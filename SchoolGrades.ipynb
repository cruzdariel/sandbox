```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("SchoolGrades24.csv")
pd.set_option('display.max_columns', 500)
```

```python
df.describe()
df.head()
```

```python
schools = ['COLONIAL HIGH',
           'LAKE NONA HIGH',
           'INNOVATION MIDDLE',
           'LAKE NONA MIDDLE',
           'LIBERTY MIDDLE',
           'ODYSSEY MIDDLE',
           'ROBERTO CLEMENTE MIDDLE',
           'ANDOVER ELEMENTARY',
           'CHICKASAW ELEMENTARY',
           'CYPRESS SPRINGS ELEMENTARY',
           'DEERWOOD ELEMENTARY',
           'EAGLE CREEK ELEMENTARY',
           'ENGELWOOD ELEMENTARY',
           'FORSYTH WOODS ELEMENTARY',
           'HIDDEN OAKS ELEMENTARY',
           'LAUREATE PARK ELEMENTARY',
           'LITTLE RIVER ELEMENTARY',
           'MCCOY ELEMENTARY',
           'MOSS PARK ELEMENTARY',
           'NORTHLAKE PARK ELEMENTARY',
           'PINAR ELEMENTARY',
           'STONE LAKES ELEMENTARY',
           'SUN BLAZE ELEMENTARY',
           'THREE POINTS ELEMENTARY',
           'VENTURA ELEMENTARY',
           'VILLAGE PARK ELEMENTARY',
           'VISTA LAKES ELEMENTARY',
           'VISTA POINTE ELEMENTARY']
df = df[df['School Name'].isin(schools)]
df
```

```python
df_grades = df[['School Name','Grade 2024', 'Informational Baseline Grade 2023','Grade 2022','Grade 2021','Grade 2019','Grade 2018','Grade 2017','Grade 2016','Informational Baseline Grade 2015','Grade 2014','Grade 2013','Grade 2012','Grade 2011','Grade 2010','Grade 2009','Grade 2008','Grade 2007','Grade 2006','Grade 2005','Grade 2004','Grade 2003','Grade 2002','Grade 2001','Grade 1999']]

rename_mapping = {
    'Grade 2024': '2024',
    'Informational Baseline Grade 2023': '2023',
    'Grade 2022': '2022',
    'Grade 2021': '2021',
    'Grade 2019': '2019',
    'Grade 2018': '2018',
    'Grade 2017': '2017',
    'Grade 2016': '2016',
    'Informational Baseline Grade 2015': '2015',
    'Grade 2014': '2014',
    'Grade 2013': '2013',
    'Grade 2012': '2012',
    'Grade 2011': '2011',
    'Grade 2010': '2010',
    'Grade 2009': '2009',
    'Grade 2008': '2008',
    'Grade 2007': '2007',
    'Grade 2006': '2006',
    'Grade 2005': '2005',
    'Grade 2004': '2004',
    'Grade 2003': '2003',
    'Grade 2002': '2002',
    'Grade 2001': '2001',
    'Grade 1999': '1999'
}

df_grades.rename(columns=rename_mapping, inplace=True)

df_grades.set_index('School Name', inplace=True)
grade_mapping = {'A': 4, 'B': 3, 'C': 2, 'D': 1, 'F': 0}
df_grades = df_grades.applymap(lambda x: grade_mapping[x] if x in grade_mapping else None)
```

```python
plt.style.use('ggplot')
for schools in df_grades.index:
  plt.plot(df_grades.columns, df_grades.loc[schools], label=schools)
  plt.title(schools)
  plt.xticks(rotation=40)
  plt.xlabel('Year')
  plt.ylabel('Grade')
  plt.show()
```

```python
df_grades['Change'] = df_grades['2024'] - df_grades['2023']
```

```python
# Negative grade change (went down)
df_grades[df_grades['Change'] < 0]
```

```python
# Positive grade change (went up)
df_grades[df_grades['Change'] > 0]
```

```python
# Neutral grade change (stayed the same)
df_grades[df_grades['Change'] == 0]
```

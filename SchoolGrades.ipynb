```python
schools = ['COLONIAL HIGH',
           'LAKE NONA HIGH',
           'INNOVATION MIDDLE',
           'LAKE NONA MIDDLE',
           'LIBERTY MIDDLE',
           'ODYSSEY MIDDLE',
           'ROBERTO CLEMENTE MIDDLE',
           'ANDOVER ELEMENTARY',
           'CHICKASAW ELEMENTARY',
           'CYPRESS SPRINGS ELEMENTARY',
           'DEERWOOD ELEMENTARY',
           'EAGLE CREEK ELEMENTARY',
           'ENGELWOOD ELEMENTARY',
           'FORSYTH WOODS ELEMENTARY',
           'HIDDEN OAKS ELEMENTARY',
           'LAUREATE PARK ELEMENTARY',
           'LITTLE RIVER ELEMENTARY',
           'MICHAEL MCCOY ELEMENTARY',
           'MOSS PARK ELEMENTARY',
           'NORTHLAKE PARK COMMUNITY ELEMENTARY',
           'PINAR ELEMENTARY',
           'STONE LAKES ELEMENTARY',
           'SUN BLAZE ELEMENTARY',
           'THREE POINTS ELEMENTARY',
           'VENTURA ELEMENTARY',
           'VILLAGE PARK ELEMENTARY',
           'VISTA LAKES ELEMENTARY',
           'VISTA POINTE ELEMENTARY']

df_regression = pd.read_csv('SchoolGrades24.csv')
df_regression = df_regression[df_regression['School Name'].isin(schools)]
df_regression.rename(columns=rename_mapping, inplace=True)
df_regression = df_regression.applymap(lambda x: grade_mapping[x] if x in grade_mapping else x)
```

```python
import statsmodels.formula.api as smf
df_regression['reg2024'] = df_regression['2024']/4
df_regression['EconDisadStu'] = df_regression['Percent of Economically Disadvantaged Students']/100
df_regression['regmathachievement'] = df_regression['Mathematics Achievement']/100
df_regression['regsciachievement'] = df_regression['Science Achievement']/100
model = smf.ols('reg2024 ~ EconDisadStu', data = df_regression)
results = model.fit()
print(results.summary())

print("      \n\n\n            ")

model = smf.ols('regmathachievement ~ EconDisadStu', data = df_regression)
results = model.fit()
print(results.summary())

print("      \n\n\n            ")

model = smf.ols('regsciachievement ~ EconDisadStu', data = df_regression)
results = model.fit()
print(results.summary())
```
